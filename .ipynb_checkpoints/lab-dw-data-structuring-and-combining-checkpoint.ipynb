```python
# 1. Combine the three datasets and then apply the cleaning function to the new combined dataset.

import pandas as pd

url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df_file1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df_file2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df_file3 = pd.read_csv(url3)
```

```python
df_merged_1_2 = pd.merge(df_file1, df_file2, on='Customer', how='left')
```

```python
df_merged_1_2
```

```python
df_merged_1_2.rename(columns={'Customer': 'customer'}, inplace=True)

df_merged_1_2.rename(columns={'ST_x': 'state'}, inplace=True)
df_merged_1_2.rename(columns={'ST_y': 'state'}, inplace=True)

df_merged_1_2.rename(columns={'GENDER_x': 'gender'}, inplace=True)
df_merged_1_2.rename(columns={'GENDER_y': 'gender'}, inplace=True)

df_merged_1_2.rename(columns={'Education_x': 'education'}, inplace=True)
df_merged_1_2.rename(columns={'Education_y': 'education'}, inplace=True)

df_merged_1_2.rename(columns={'Customer Lifetime Value_x': 'customer_lifetime_value'}, inplace=True)
df_merged_1_2.rename(columns={'Customer Lifetime Value_y': 'customer_lifetime_value'}, inplace=True)

df_merged_1_2.rename(columns={'Income_x': 'income'}, inplace=True)
df_merged_1_2.rename(columns={'Income_y': 'income'}, inplace=True)

df_merged_1_2.rename(columns={'Monthly Premium Auto_x': 'monthly_premium_auto'}, inplace=True)
df_merged_1_2.rename(columns={'Monthly Premium Auto_y': 'monthly_premium_auto'}, inplace=True)

df_merged_1_2.rename(columns={'Number of Open Complaints_x': 'number_of_open_complaints'}, inplace=True)
df_merged_1_2.rename(columns={'Number of Open Complaints_y': 'number_of_open_complaints'}, inplace=True)

df_merged_1_2.rename(columns={'Policy Type_x': 'policy_type'}, inplace=True)
df_merged_1_2.rename(columns={'Policy Type_y': 'policy_type'}, inplace=True)

df_merged_1_2.rename(columns={'Vehicle Class_x': 'vehicle_class'}, inplace=True)
df_merged_1_2.rename(columns={'Vehicle Class_y': 'vehicle_class'}, inplace=True)

df_merged_1_2.rename(columns={'Total Claim Amount_x': 'total_claim_amount'}, inplace=True)
df_merged_1_2.rename(columns={'Total Claim Amount_y': 'total_claim_amount'}, inplace=True)

df_merged_1_2
```
